```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
import numpy as np
import matplotlib.pyplot as plt
from argparse import ArgumentParser
import torchvision.models as models
from tqdm import tqdm
```

```python
def str2bool(v):
    if v.lower() == "true":
        return True
    return False
```

```python
parser = ArgumentParser()
parser.add_argument("--model", type=str, default="vgg")
parser.add_argument("--use_cuda", type=str2bool, default=True)
use_cuda=True
args = parser.parse_args(args=[])
print(args)
```

```python
use_cuda = args.use_cuda
```

```python
# Define what device we are using
print("CUDA Available: ",torch.cuda.is_available())
device = torch.device("cuda" if (use_cuda and torch.cuda.is_available()) else "cpu")
```

```python
# Initialize the network
if args.model == "LeNet":
    model = LeNet().to(device)
elif args.model == "vgg":
    in_channels = 1
    cfgs = {
        'VGG_gray': [64, 'M', 128, 'M'],
    }
    model = VGG(in_channels, cfgs["VGG_gray"]).to(device)
else:
    raise RuntimeError("No such model")
```

```python
model(torch.randn(10, 1, 28, 28).to(device))
```

```python
train_loader = torch.utils.data.DataLoader(
    datasets.MNIST('../data', train=True, download=True, transform=transforms.Compose([
            transforms.ToTensor(),
            ])), 
        batch_size=32, shuffle=True)

test_loader = torch.utils.data.DataLoader(
    datasets.MNIST('../data', train=False, download=True, transform=transforms.Compose([
            transforms.ToTensor(),
            ])), 
        batch_size=1, shuffle=True)
```

```python
def train(model, device, train_loader, optimizer, loss_f, epoch):
    model.train()
    for batch_idx, (features, label) in enumerate(train_loader):
        
        features, label = features.to(device), label.to(device)
        optimizer.zero_grad()
        output = model(features)
        loss = loss_f(output, label)
        loss.backward()
        optimizer.step()
        
        if batch_idx % 10 == 0:
            print("\rloss: " + str(loss.item()), end="")
            
            
def test(model, device, test_loader, loss_f):
    
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for batch_idx, (features, label) in enumerate(test_loader):

            features, label = features.to(device), label.to(device)
            output = model(features)
            loss = loss_f(output, label)
            test_loss += loss.item()

            pred = output.argmax(dim=1, keepdim=True)  # get the index of the max log-probability
            correct += pred.eq(label.view_as(pred)).sum().item()
            
            if batch_idx % 10 == 0:
                print("\rloss: " + str(loss.item()), end="")
    
    test_loss /= len(test_loader.dataset)
    correct /= len(test_loader.dataset)
    
    print("\ntest_loss:", test_loss)
    print("accuracy:", correct)
    
```

```python
epochs = 3
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, betas=(0.9, 0.999), eps=1e-08, weight_decay=0, amsgrad=False)
loss_f = nn.CrossEntropyLoss()

for epoch in range(epochs):
    train(model, device, train_loader, optimizer, loss_f, epoch)
    test(model, device, test_loader, loss_f)
```

```python
epsilons = [0, .05, .1, .15, .2, .25, .3]
```

```python
# https://pytorch.org/tutorials/beginner/fgsm_tutorial.html

accus = []
examples = []
for epsilon in tqdm(epsilons):
    accuracy, example = test_fgsm_attack(model, device, test_loader, loss_f, epsilon)
    accus.append(accuracy)
    examples.append(example)
```

```python
# https://pytorch.org/tutorials/beginner/fgsm_tutorial.html

plt.figure(figsize=(5,5))
plt.plot(epsilons, accus, "*-")
plt.yticks(np.arange(0, 1.1, step=0.1))
plt.xticks(np.arange(0, .35, step=0.05))
plt.title("Accuracy vs Epsilon")
plt.xlabel("Epsilon")
plt.ylabel("Accuracy")
plt.show()
```

```python
# https://pytorch.org/tutorials/beginner/fgsm_tutorial.html

# Plot several examples of adversarial samples at each epsilon
cnt = 0
plt.figure(figsize=(8,10))
for i in range(len(epsilons)):
    for j in range(len(examples[i])):
        cnt += 1
        plt.subplot(len(epsilons),len(examples[0]),cnt)
        plt.xticks([], [])
        plt.yticks([], [])
        if j == 0:
            plt.ylabel("Eps: {}".format(epsilons[i]), fontsize=14)
        orig,adv,ex = examples[i][j]
        plt.title("{} -> {}".format(orig, adv))
        plt.imshow(ex, cmap="gray")
plt.tight_layout()
plt.show()
```
